$\alpha$ -- коэффициент влияния предыдущего элемента временного ряда на предсказание. При $\alpha$ близкой к 1 мы теряем адаптивную способность и начинает выдавать тот же самый ответ. 

**Вопрос 2**: Каким следует выбрать параметр сглаживания тренда $\alpha_2$ в модели Хольта (линейный тренд) в случае, когда вы предсказываете временной ряд 1) с плавно меняющимя трендом; 2) стохастически меняющися трендом?

Надеюсь, под $\alpha_2$ имелось в виду тот коэффициент, что есть в формуле <a href=http://www.machinelearning.ru/wiki/index.php?title=Модель_Хольта>тут</a>

Если $\alpha_2 = 0$, то $\forall t: \beta_t = const$. По-этому модель станет ${\hat{y}}_{t+d} = a_t + d \cdot \beta_0 = a_t$ -- экспоненциальное сглаживание, которое подойдет для стохастически меняющегося тренда (его мат. ожидание будет постоянным). Напротив, если $\alpha_2 >> 0$, то мы оцениваем плавный тренд

**Вопрос 3**: Докажите  эквивалетность формул расчёта (в рекуррентной форме и в форме корректировки прогноза (correction error form) )скрытых компонент временного ряда в модели Тейла-Вейджа (период сезонность обозначается $p$):
$$l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} ) \color{red}{=} l_{t-1} + b_{t-1} + \alpha e_t;$$,
    $$ b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} \color{red}{=} b_{t-1} + \alpha\beta e_t;$$,
    $$ s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} \color{red}{=} s_{t-p} + \gamma(1-\alpha)e_t.$$,



$$ l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} ) = \alpha(y_t - s_{t-p} - l_{t-1} - b_{t-1}) + l_{t-1} + b_{t-1} = \alpha e_t + l_{t-1} + b_{t-1},$$
Здесь воспользовались определением $e_t = y_t - \hat{y}_t = y_t - s_{t-p} - l_{t-1} - b_{t-1}$
    Остальные равенства проверяются раскрытием скобок и приведением подобных слагаемых при $\alpha\beta$ и $\gamma(1-\alpha)$, а также заменой $l_t$ на только что полученную формулу.

**Вопрос 4**: Выведите формулу расчёта скрытых компонент в форме корректировки прогноза (correction error form) временного ряда в модели ЭС с линейным затухающим трендом (WS with additive damped trend):

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$,
   
Скрытые компоненты ВР согласно модели в рекуррентной форме записываются:
    $$ l_{t}       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right),$$,

```python
# Start with this code
###################### Winters Exponential Smoothing #########################
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha <scalar in [0,1]> - smoothing parameter
#    delts <scalar in [0,1]> - seasonality smoothing parameter

def WintersExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    delta = Params['delta']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN] * (T + h)
    
    l = np.NaN  # initialize ts level 
    s = [np.NaN] * p # initalize seasonality values (it must be vector of lenth p)
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l = 0
 
            if math.isnan(s[cntr % p]):
                s[cntr % p]= 0 # initialize 
 
            l_last = l
            l = alpha * (x[cntr] - s[cntr % p]) + (1 - alpha) * l
            s[cntr % p] = delta * (x[cntr] - l_last) + (1 - delta) * s[cntr % p]
            
        FORECAST[cntr+h] = l + s[(cntr+h) % p]
    return FORECAST
```

```python
def Theil_Wage_Model(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    delta = Params['delta']
    p = Params['seasonality_period']
    beta = Params['beta']
    
    FORECAST = [np.NaN] * (T + h)
    
    l = np.NaN  # initialize ts level 
    s = [np.NaN] * p # initalize seasonality values (it must be vector of lenth p)
    b = np.NaN 
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l = 0
 
            if math.isnan(s[cntr % p]):
                s[cntr % p] = 0
            
            if math.isnan(b):
                b = 0
            
            l_last = l
            b_last = b
            
            l = alpha * (x[cntr] - s[cntr % p]) + (1 - alpha) * (l_last + b_last)
            s[cntr % p] = delta * (x[cntr] - l_last) +  (1 - delta) * s[cntr % p]
            b = beta * (l - l_last) + (1 - beta) * b_last
            
        FORECAST[cntr + h] = l + s[(cntr + h) % p]
    return FORECAST
```

```python
def Winters_Model(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    beta = Params['beta']
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN] * (T + h)
    
    l = np.NaN  # initialize ts level 
    s = [np.NaN] * p # initalize seasonality values (it must be vector of lenth p)
    b = np.NaN 
    
    for cntr in range(T):
        if not math.isnan(x[cntr]):
            if math.isnan(l):
                l = 1
 
            if math.isnan(s[cntr % p]):
                s[cntr % p] = 1
            
            
            l_prev = l
            l = alpha * (x[cntr] / s[cntr % p]) + (1 - alpha) * l_prev
            s[cntr % p] = beta * (x[cntr] / l_prev) + (1 - beta) * s[cntr % p]
            
        FORECAST[cntr + h] = l * s[(cntr+h) % p]
    return FORECAST
```
