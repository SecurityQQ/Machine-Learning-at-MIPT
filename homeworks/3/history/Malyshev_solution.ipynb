```python
def transform(data):
    for i, row in data.iterrows():
        shift = row['shift']
        if 1 <= row['week'] - shift <= 52:
            data.set_value(i, 'week', row['week'] - shift)
        else:
            if (row['year'] == 2015):
                data.set_value(i, 'week', 53 + (row['week'] - shift))
            else:
                data.set_value(i, 'week', 52 + (row['week'] - shift))
            data.set_value(i, 'year', row['year'] - 1)
```

```python
def smape100(estimator, X, Y):
    pred = estimator.predict(X)
    return 100 * np.sum(2 * np.abs(pred - Y) / (np.abs(pred) + np.abs(Y)))/len(pred)
```
